```python
import numpy as np

%matplotlib inline
import matplotlib.pylab as plt
```

```python
#
# empirical check of the CTI
#
# Al Ka from PANTHER facility, data read from the plot in M. Freyberg presentation in Mallorca, Feb 2005, page 11
#
modes = ["FF","LW","SW"]
alk0 = 262.0
#alk0 = np.array([262,264,263],dtype=float)
alk190 = np.array([226,233,230],dtype=float)
al_ratio = alk190/alk0
#
# Cu Ka
#
cuka0 = 1435.0
#cuka0 = np.array([1435,1435,1455],dtype=float)
cuka190 = np.array([1315,1330,1330],dtype=float)
cuka_ratio = cuka190/cuka0
#
# Cu L
#
cul0 = 161.0
cul190 = np.array([134,140,139])
cul_ratio = cul190/cul0
#
# O-K
#
ok0 = 90.0
ok190 = np.array([71,75,75])
ok_ratio = ok190/ok0
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.subplots()
for i,xmode in enumerate(modes):
    ax.semilogx([ok0,cul0,alk0,cuka0],[ok_ratio[i],cul_ratio[i],al_ratio[i],cu_ratio[i]],'-o',label=f'{xmode}')
    #ax.plot([ok0,cul0,alk0,cuka0],[ok_ratio[i],cul_ratio[i],al_ratio[i],cu_ratio[i]],'-o',label=f'{xmode}')
for j in zip([ok0,cul0,alk0,cuka0],["O-K","Cu-L","Al-K","Cu-K"],[ok_ratio[0],cul_ratio[0],al_ratio[0],cu_ratio[0]]): 
    ax.text(j[0],j[2]-0.005,j[1])
ax.set_xlabel("Log10(Line peak (ADU))")
ax.set_ylabel("Line peak (RAWY=190)/(RAWY=0)")
ax.grid(True)
ax.set_title("Energy dependence at boresight, pinhole measurements, PANTHER 2004/2005")
ax.legend();
wdir = "/xdata/xcaldata/XMM/IVAN/"
plt.savefig(f'{wdir}/panther_energy_deps_plot.png',dpi=100)
plt.show()
plt.close()
```

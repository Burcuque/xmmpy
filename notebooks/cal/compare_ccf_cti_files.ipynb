# Compare two EPN_CTI CCF files

The CCF files ready for SAS integration are available in this folder: `/ccf/priv/dt`, my CCF development folder is `/xdata/xcaldata/XMM/IVAN/ccfdev`.

```python
import os, sys
import numpy as np
#from astropy.io import fits
from astropy.table import setdiff, Table
import difflib

dir1 = '/ccf/priv/dt'
dir2 = '/xdata/xcaldata/XMM/IVAN/ccfdev'
```

```python
#
# read the CCF files
#
extname = 'LONG_TERM_CTI'
hdu1 = Table.read(f"{dir1}/EPN_CTI_0049.CCF",hdu=extname)
n1 = len(hdu1)
hdu2 = Table.read(f"{dir1}/EPN_CTI_0050.CCF",hdu=extname)
n2 = len(hdu2)
hdu3 = Table.read(f"{dir2}/EPN_CTI_0049.CCF_test53",hdu=extname)
n3 = len(hdu3)
```

```python
#
# check the lengths
#
if (n1 != n2 or n2 != n3):
    print (f"Different lengths of LONG_TERM_CTI the tables: {n1}, {n2}, {n3}")
else:
    print (f"Same number of elements in LONG_TERM_CTI the tables: {n1}, {n2}, {n3}")
```

```python
#
# now compare line by line using difflib
#
for j in np.arange(n1):
    line1 = f"{hdu1.as_array()[j]}".split()
    line2 = f"{hdu2.as_array()[j]}".split()
    line3 = f"{hdu3.as_array()[j]}".split()
    diff23 = difflib.context_diff(line2,line3)
    sys.stdout.writelines(diff23)    
#
```

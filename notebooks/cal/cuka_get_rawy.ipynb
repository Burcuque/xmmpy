# Deriving the average RAWY

Using the event lists, selecting the events with mask and derive an average RAWY for each CCD.

```python
import os
import numpy as np
import pandas as pd
from astropy.table import Table
home = os.path.expanduser('~')
wdir = home + '/IVAN/Cu-line/RAWY'
```

```python
t = Table.read(f'{wdir}/all_rawy_results.csv')
df = t.to_pandas()
boxplot = df.boxplot(column=['rawy'],by='ccd',figsize=(10,5))
fig = boxplot.get_figure()
fig.savefig(f"{wdir}/boxplot_rawy_all.png",dpi=100)
```

```python
grouped = df.groupby('ccd')
grouped['rawy'].describe()
```

```python
a = grouped['rawy'].describe()
```

```python
df = pd.read_csv(f'{wdir}/ao18_otac_by_id.csv')
```

```python
#
categories = {
  'A': 'Stars, White Dwarfs and Solar System',
  'B': 'White Dwarf Binaries, Neutron Star Binaries, Cataclysmic Variables, ULXs and Black Holes',
  'C': 'Supernovae, Supernova Remnants, Diffuse (galactic) Emission and Isolated Neutron Stars',
  'E': 'Active Galactic Nuclei, Quasars, BL-Lac Objects',
  'F': 'Galaxies, Groups of Galaxies, Clusters of Galaxies and Superclusters',
  'G': 'Cosmology, Extragalactic Deep Fields and Large Extragalactic Areas'}
#
done_abs = {}
#
# outut file as markDown
#
fout = open(f'{wdir}/ao18_abstracts.md','w')
#
for xcat in categories.keys():
    df_sel = df.loc[df['cat'] == xcat]
    ntot = len(df_sel)
    print (f'# {categories[xcat]}, N = {ntot}')
    print ()
    print (f'# {categories[xcat]}',file=fout)
    print (file=fout)
    nobs = len(df_sel['obsid'])
    #
    # get a list of targets as I am stupid and cannot do it in one loop
    #
    target = {}
    #
    for j in range(nobs):
        iobs = df_sel['obsid'].iloc[j]
        sobs = f"{iobs:08}"
        xobs = sobs[0:6]
        if (xobs in target.keys()):
            target[xobs] += f"{df_sel['target'].iloc[j]}, "
        else:
            target[xobs] = f"{df_sel['target'].iloc[j]}, "
    #
    jk = 1
    for j in range(nobs):
        iobs = df_sel['obsid'].iloc[j]
        sobs = f"{iobs:08}"
        xobs = sobs[0:6]
        if (xobs in done_abs.keys()):
            continue
        #
        abs_file = f'{wdir}/ao18_abstracts/{xobs}_PROP.asc'
        if (os.path.isfile(abs_file)):
            #print (f"found an abstract file {abs_file}")
            with open(abs_file,'r') as qf:
                lines = qf.readlines()
            print (f"## {xcat}{jk}: **PI:** {df_sel['pi'].iloc[j]}, OBSID: {xobs}\n",file=fout)
            print (f"**Title:** {lines[0]}",file=fout)
            merged = ""
            for k in lines[1:]:
                merged += (k.rstrip() + " ")
            print (f"**Abstract:** {merged}\n",file=fout)
            done_abs[xobs] = True
            jk += 1
        else:
            print (f"Abstract file {abs_file} not found")
        ntargets = len(target[xobs].split(',')) - 1
        print (f"**Targets:** (n={ntargets}) {target[xobs][:-2]}",file=fout)
        print (file=fout)
fout.close()
```

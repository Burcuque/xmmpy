# Zoouniverse stats from the Hunt for Galaxy Clusters project

[The Hunt for Galaxy Clusters](https://www.zooniverse.org/projects/matej-dot-kosiba/the-hunt-for-galaxy-clusters) zoouniverse project finished in April 2019.

```python
import os
import json
import pandas as pd
```

```python
wdir = os.path.join(os.path.expanduser('~'),'Desktop','HuntForClusters')
```

```python
zoofile = f'{wdir}/the-hunt-for-galaxy-clusters-classifications_24-June-2019.csv.gz'
#zoofile = os.path.expanduser('~') + '/Downloads/the-hunt-for-galaxy-clusters-subjects.csv'
df = pd.read_csv(zoofile)
```

```python
df = df.query('workflow_id == 7298')
df[-20:-10]
```

```python
df.iloc[10]['user_id']
```

```python
ntot = len(df)
tasks = ['T0', 'T1', 'T2', 'T3', 'T4', 'T5','T6', 'T7']
```

```python
output = []
for j in range(ntot):
    if (j % 1000 == 0):
        print (f"Doing {j+1}/{ntot}")
    result = []
    xk = json.loads(df.iloc[j]['subject_data'])
    uid = df.iloc[j]['user_id']
    uname = df.iloc[j]['user_name']
    for k in xk.keys():
        xid = json.loads(df.iloc[j]['subject_data'])[k]['subject_id']
    #print (xid)
    #
    result.append(xid)
    result.append(uid)
    result.append(uname)
    #
    xann = json.loads(df.iloc[j]['annotations'])
    nann = len(xann)
    xtasks = {'T0':'', 'T1':'', 'T2':'', 'T3':'', 'T4':'', 'T5':'','T6':'', 'T7':''}
    for it in range(nann):
        xtasks[xann[it]['task']] = xann[it]['value']
    [result.append(xtasks[k]) for k in xtasks.keys()]
    output.append(result)
#
```

```python
outdf = pd.DataFrame(output,columns=['ID','UID','UNAME','T0', 'T1', 'T2', 'T3', 'T4', 'T5','T6', 'T7'])
outdf.to_pickle(f'{wdir}/output_table.pkl')
```

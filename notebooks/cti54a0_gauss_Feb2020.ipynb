```python
def read_cti_ccf(ccf_file,mode_id=2, energy_index=0):
    #
    # utility function to read the EPN CTI calfile and get the dat ain a dictionary
    #
    # mode_id = 0 for PRIME_FULL_WINDOW
    # mode_id = 1 for PRIME_FULL_WINDOW_EXTENDED
    # mode_id = 2 for PRIME_LARGE_WINDOW
    # mode_id = 3 for PRIME_SMALL_WINDOW
    #
    # energy_index = 0 is for Al Ka at 1.486
    # energy_index = 1 is for Mn Ka at 5.8988
    # energy_index = 2 is for Fe Ka at 6.4 keV
    #
    try:
        t = Table.read(f'{ccf_file}',hdu='LONG_TERM_CTI')
        tmp = Table.read(f'{ccf_file}',hdu='LTC_TIMES')
        #times = tmp.data['TIME'][0].flatten()                           
    except:
        print (f"Cannot read CCF file {ccf_file}, it must contain extension with name \"LONG_TERM_CTI\"")
        return None
    times = tmp['TIME'][0]
    tx = t.group_by('MODE_ID')
    mmask = tx.groups.keys['MODE_ID'] == mode_id
    tenergy = tx.groups[mmask].group_by('ENERGY')
    ixxx = np.arange(len(tenergy.groups))
    if (energy_index not in ixxx):
        print (f"The index for the energy {energy_index} is not available for this mode. Only indices: ", ixxx)
        return None
    grp = tenergy.groups[energy_index]
    #print (grp)
    tout = {}
    if (mode_id != 3):
        for iccd in np.arange(1,13):
            xmask = grp['CCD_ID'] == iccd
            tcoeff = grp[xmask]['T_COEFF']
            tout[iccd] = {'mode_id': mode_id, 'ccd': iccd, 'energy': grp['ENERGY'][0],
                'times': times, 'tcoeff': tcoeff[0]}
    else:
        tcoeff = grp['T_COEFF']
        tout[0] = {'mode_id': mode_id, 'ccd': 4, 'energy': grp['ENERGY'][0],
            'times': times, 'tcoeff': tcoeff[0]}
    return tout
```

## Derive new CTI curve, per CCD

```python
def make_cti_per_ccd(ccd,dataframe,cti_tab, plot_it=False, plot_file=None):
    # 
    # Purpose:
    #    Derive a new long-term CTI curve per CCD
    # Inputs:
    #    ccd - the CCD number (from 1 to 12), do not use the central CCDs as the Cu Kalpha is not prominent there.
    #    dataframe - the pandas dataframe with the Cu Kalpha fit results (after running cu_line_analysis_ctix.py)
    #    cti_tab - the current/testing CTI CCF file, will use the CTI curves from this file up to t=14 and change them for t > 14
    #    plot_it - if set, will plot the results.
    #    if plot_file is set, then the plotting results will be saved to this file.
    # Output:
    #    the derived long-term CTI curve (Eobs/Elab) for the ccd
    # 
    # Method:
    #    1. Use the long-term CTI curve in cti_tab up to t=14
    #    2. Linearly extrapolate for t > 14 using the slope from t=10 to t=14.
    #    3. Derive the residuals of the observed Eobs/Elab with respect to the extrapolated curve
    #    4. Fit a Gaussian1D model to the residual, the mean is fixed to 20 and the st.dev. fixed to 2.4. Only fit for amplitude.
    #    5. The new Eobs/Elab is the combined curve from the CTI file (t<14) and the Gaussian fit added to the residual.
    #
    #
    if (ccd not in range(1,13)):
        print (f"CCD number must be one of [1,2,...,12], your input is {ccd}. Cannot continue.")
        return None
    #
    xtab = dataframe[(dataframe.ccd == ccd) & (dataframe.ratioE <= 1.0)]
    #
    xin = xtab.delta_time
    yin = xtab.ratioE
    yin_err = xtab.ratioE_err
    #
    # extrapolate with the same slope as the one from t=10 to t=14
    istart = 10
    iend = 14
    # 
    #
    newy = cti_tab[ccd]['y'].copy()
    m = (newy[iend] - newy[istart])/(cti_tab[ccd]['x'][iend] - cti_tab[ccd]['x'][istart])
    newy[iend:] = m*(cti_tab[ccd]['x'][iend:] - cti_tab[ccd]['x'][iend]) + newy[iend]
    intercept = -m*cti_tab[ccd]['x'][iend] + newy[iend]
    #
    # now calculate the residuals
    #
    cuKa0 = 8040.0 # lab energy of the Cu Kalpha line in eV
    fitted = np.interp(xin,cti_tab[ccd]['x'],newy)
    residual = (yin - fitted)*cuKa0 # in eV
    residual_err = yin_err*cuKa0
    #w = np.ones_like(residual)
    w = 1.0/residual_err
    w.loc[(~np.isfinite(w))] = 0.0
    #
    # fit a Gaussian to the residual
    #
    fitter = fitting.SLSQPLSQFitter()
    #
    # fixing the peak and sigma of the gaussian for all CCDs, hardcoded!
    #
    g_init = models.Gaussian1D(mean=20.0, amplitude=0.002, stddev=2.4)
    # keep only the amplitude as free parameter
    g_init.mean.fixed = True
    g_init.stddev.fixed = True
    g_fit = fitter(g_init, xin, residual, weights=w)
    gauss = g_fit(cti_tab[ccd]['x'])
    #
    #
    # new residual
    #
    newfit = newy+ gauss/cuKa0
    #
    fitted2 = np.interp(xin,cti_tab[ccd]['x'],newfit)
    residual2 = (yin - fitted2)*cuKa0 # in eV
    residual2_err = yin_err*cuKa0
    #
    if (plot_it):
        fig, ax = plt.subplots(3,1,figsize=(10,10),sharex=True)
        zoomx = (0,25)
        zoomy = (0.965,1.0)
        #
        ax[0].plot(cti_tab[ccd]['x'],newy,color='red',ls='dashed',lw=3,label='CCFv52 + linear')
        # and the data
        ax[0].errorbar(xin,yin,yerr=(yin_err,yin_err),fmt='o',label=f'CCD #{ccd:02}',alpha=0.5)
        ax[0].set_xlim(zoomx)
        ax[0].set_ylim(zoomy)
        ax[0].grid(True)
        ax[0].legend(loc=3)
        ax[0].set_title(fr"Cu K$\alpha$ fit for EPIC-PN Large Window Mode")
        ax[0].set_ylabel(r"E$_{obs}$/E$_{lab}$")
        ax[0].set_xlabel("")
        #
        # the residual plot
        #
        ax[1].errorbar(xin,residual,yerr=residual_err,fmt='o',label=f'CCD #{ccd:02}',alpha=0.5)
        ax[1].plot(cti_tab[ccd]['x'],gauss,color='green',ls='dotted',lw=3.0)
        ax[1].axhline(0.0,lw=3,ls='dashed',color='black')
        ax[1].set_xlim(zoomx)
        ax[1].grid(True)
        ax[1].legend(loc=3)
        ax[1].set_ylabel("Residual #1")
        ax[0].plot(cti_tab[ccd]['x'],newfit,color='green',ls='dotted',lw=3,label='Combined')
        #
        ax[2].errorbar(xin,residual2,yerr=residual2_err,fmt='o',label=f'CCD #{ccd:02}',alpha=0.5)
        ax[2].axhline(0.0,lw=3,ls='dashed',color='black')
        ax[2].set_xlim(zoomx)
        ax[2].set_ylim((-50,50))
        ax[2].axhline(-20.0,lw=3,ls='dotted',color='green')
        ax[2].axhline(20.0,lw=3,ls='dotted',color='green')
        ax[2].grid(True)
        ax[2].legend(loc=3)
        #ax[2].set_title(fr"Cu K$\alpha$ fit for EPIC-PN Large Window Mode")
        ax[2].set_ylabel("Residual #2")
        ax[2].set_xlabel("Time since 2000-01-01T00:00:00 (years)")
        #
        # some stats on the final residual
        #
        med1, low1, up1 = np.quantile(residual2,[0.5,0.16,0.84])
        ax[2].text(0.1,0.8,f'med: {med1:.1f} ({low1:.1f},{up1:.1f}) eV',transform=ax[2].transAxes)
        plt.tight_layout()
        plt.show()
        if (plot_file != None):
            plt.savefig(plot_file,dpi=100)
            plt.close()
    return newfit
```

```python
def make_cti_per_quad(quad,dataframe,cti_tab, plot_it=False, plot_file=None):
    # 
    # Purpose:
    #    Derive a new long-term CTI curve per quadrant
    # Inputs:
    #    quadrant - the quadrant number (from 0 to 3)
    #    dataframe - the pandas dataframe with the Cu Kalpha fit results (after running cu_line_analysis_ctix.py)
    #    cti_tab - the current/testing CTI CCF file, will use the CTI curves from this file up to t=14 and change them for t > 14
    #    plot_it - if set, will plot the results.
    #    if plot_file is set, then the plotting results will be saved to this file.
    # Output:
    #    the derived long-term CTI curve (Eobs/Elab) for the quadrant, not using the central CCDs where the CuKalpha is not prominent
    # 
    # Method:
    #    1. Use the long-term CTI curve in cti_tab up to t=14
    #    2. Linearly extrapolate for t > 14 using the slope from t=10 to t=14.
    #    3. Derive the residuals of the observed Eobs/Elab with respect to the extrapolated curve
    #    4. Fit a Gaussian1D model to the residual, the mean is fixed to 20 and the st.dev. fixed to 2.4. Only fit for amplitude.
    #    5. The new Eobs/Elab is the combined curve from the CTI file (t<14) and the Gaussian fit added to the residual.
    #
    #
    #
    if (quad not in [0,1,2,3]):
        print (f"Quandrant must be one of [0,1,2,3], your input is {quad}. Cannot continue.")
        return None
    quad = f'q{quad}'
    #
    ccds = {'q0': [1,2,3], 'q1': [4,5,6], 'q2': [7,8,9], 'q3': [10,11,12]}
    #
    xtab = dataframe[(dataframe.ccd == ccds[quad][1]) | (dataframe.ccd == ccds[quad][2]) & (dataframe.ratioE <= 1.0)]
    #
    xin = xtab.delta_time
    yin = xtab.ratioE
    yin_err = xtab.ratioE_err
    #
    # extrapolate with the same slope as the one from t=10 to t=14
    istart = 10
    iend = 14
    # 
    ccd = ccds[quad][-1]
    #
    newy = cti_tab[ccd]['y'].copy()
    m = (newy[iend] - newy[istart])/(cti_tab[ccd]['x'][iend] - cti_tab[ccd]['x'][istart])
    newy[iend:] = m*(cti_tab[ccd]['x'][iend:] - cti_tab[ccd]['x'][iend]) + newy[iend]
    intercept = -m*cti_tab[ccd]['x'][iend] + newy[iend]
    #
    # now calculate the residuals
    #
    cuKa0 = 8040.0 # Elab in eV
    fitted = np.interp(xin,cti_tab[ccd]['x'],newy)
    residual = (yin - fitted)*cuKa0 # in eV
    residual_err = yin_err*cuKa0
    #w = np.ones_like(residual)
    w = 1.0/residual_err
    # avoid zero errors leading to inf weights
    w.loc[(~np.isfinite(w))] = 0.0
    # set zero weight for large residuals, greater than 50 eV
    w.loc[(residual.abs() >= 50.0)] = 0.0
    #
    # fit a Gaussian to the residual
    #
    fitter = fitting.SLSQPLSQFitter()
    #
    # fixing the peak and sigma of the gaussian for all CCDs, hardcoded!
    #
    g_init = models.Gaussian1D(mean=20.0, amplitude=0.002, stddev=2.4)
    # keep only the amplitude as free parameter
    g_init.mean.fixed = True
    g_init.stddev.fixed = True
    g_fit = fitter(g_init, xin, residual, weights=w)
    gauss = g_fit(cti_tab[ccd]['x'])
    #
    #
    # new residual
    #
    newfit = newy+ gauss/cuKa0
    #
    fitted2 = np.interp(xin,cti_tab[ccd]['x'],newfit)
    residual2 = (yin - fitted2)*cuKa0 # in eV
    residual2_err = yin_err*cuKa0
    #
    if (plot_it):
        fig, ax = plt.subplots(3,1,figsize=(10,10),sharex=True)
        zoomx = (0,25)
        zoomy = (0.965,1.0)
        #
        ax[0].plot(cti_tab[ccd]['x'],newy,color='red',ls='dashed',lw=3,label='CCFv52 + linear')
        # and the data
        ax[0].errorbar(xin,yin,yerr=(yin_err,yin_err),fmt='o',label=f'{quad}',alpha=0.5)
        ax[0].set_xlim(zoomx)
        ax[0].set_ylim(zoomy)
        ax[0].grid(True)
        ax[0].legend(loc=3)
        ax[0].set_title(fr"Cu K$\alpha$ fit for EPIC-PN Large Window Mode")
        ax[0].set_ylabel(r"E$_{obs}$/E$_{lab}$")
        ax[0].set_xlabel("")
        #
        # the residual plot
        #
        ax[1].errorbar(xin,residual,yerr=residual_err,fmt='o',label=f'{quad}',alpha=0.5)
        ax[1].plot(cti_tab[ccd]['x'],gauss,color='green',ls='dotted',lw=3.0)
        ax[1].axhline(0.0,lw=3,ls='dashed',color='black')
        ax[1].set_xlim(zoomx)
        ax[1].grid(True)
        ax[1].legend(loc=3)
        ax[1].set_ylabel("Residual #1")
        ax[0].plot(cti_tab[ccd]['x'],newfit,color='green',ls='dotted',lw=3,label='Combined')
        #
        ax[2].errorbar(xin,residual2,yerr=residual2_err,fmt='o',label=f'{quad}',alpha=0.5)
        ax[2].axhline(0.0,lw=3,ls='dashed',color='black')
        ax[2].axhline(-20.0,lw=3,ls='dotted',color='green')
        ax[2].axhline(20.0,lw=3,ls='dotted',color='green')
        ax[2].set_xlim(zoomx)
        ax[2].grid(True)
        ax[2].legend(loc=3)
        ax[2].set_ylim((-50,50))
        #ax[2].set_title(fr"Cu K$\alpha$ fit for EPIC-PN Large Window Mode")
        ax[2].set_ylabel("Residual #2")
        ax[2].set_xlabel("Time since 2000-01-01T00:00:00 (years)")
        #
        # some stats on the final residual
        #
        med1, low1, up1 = np.quantile(residual2,[0.5,0.16,0.84])
        ax[2].text(0.1,0.8,f'med: {med1:.1f} ({low1:.1f},{up1:.1f}) eV',transform=ax[2].transAxes)
        plt.tight_layout()
        if (plot_file != None):
            plt.savefig(plot_file,dpi=100)
            plt.show()
            plt.close()
    return newfit
```

## Prepare for the CCF

This step require the average RAWY per CCD and then convert the ratio `Q=Eobs/Elab` to the $g(t)$  function:

$Q = E_{obs}/E_{lab} = \left[\frac{1 - g(t)}{1-a_0}\right]^{RAWY}$,

where $a_0 = g(t_0)$, the measured CTI at launch. We know $a_0$ at 5.8988 keV but not at 8.04 keV. so we use the value for 5.8988 keV, $a_0 = 0.00043236$.

Rewriting the equation we obtain:

$g(t) = 1 - (1-a_0)\times Q^{1/RAWY}$.  

The tabulated function $g(t)$ is stored in the EPN_CTI calibartion file.

Will also investigate an alternative way, when $a_0$ is unknown. Then we can set it to zero:

$Q = \left(1 - g(t)\right)^{RAWY}$,

and consequently 

$g(t) = 1 - Q^{1/RAWY}$.

As shown in the boxplot above, for all non-central CCDs the average `RAWY` is 143 +/- 1 and we assume this value for the calculation of $g(t)$.

We construct `EPN_CTI_0054.CCF` with the per-quadrant curves.

```python
#
# results dictionary per CCD is the fitted Eobs/Elab (the red line).
#
# per quadrant mapping
quadrant = {1: "q0", 2: "q0", 3: "q0", 4: "q1", 5: "q1", 6: "q1",
          7: "q2", 8: "q2", 9: "q2", 10: "q3", 11: "q3", 12: "q3"}
#
# TCOEFF ( = gt1) will be calculated for a fixed RAWY and a0 for each CCDs
gt1 = {}
mean_rawy = 143.0
a0 = 0.0004
for i in range(12):
    ccd = i + 1
    iq = quadrant[ccd] # quadrants mapping
    gt1[ccd] = 1.0 - (1.0 - a0)*np.power(cti_res[iq],1.0/mean_rawy)
    gt1[ccd][0] = a0
#
```

One way to check is with plotting, another way is to dump the content to an ascii files and do `diff` from the OS.


```python
def dump_cti(ccf_file,fileout=None):
    #
    # will dump the LONG_TERM_CTI extension in an ascii file with format following the standard one
    #
    hdu = fits.open(ccf_file)
    extname = 'LONG_TERM_CTI'
    ltc = hdu[extname]
    nt = len(ltc.data)
    if (fileout != None):
        fout = open(fileout,'w')
    #
    for j in np.arange(nt):
        if (fileout != None):
            print ("{:3} : {:3} :   {:6.4f} : ".format(ltc.data['MODE_ID'][j],ltc.data['CCD_ID'][j],\
               ltc.data['ENERGY'][j]),end="",file=fout)
        #print ("{:3} : {:3} :   {:6.4f} : ".format(ltc.data['MODE_ID'][j],ltc.data['CCD_ID'][j],\
        #       ltc.data['ENERGY'][j]),end="",file=fout)
        tcoeff = ltc.data['T_COEFF']
        nrow,ncol = tcoeff.shape
        for l in np.arange(ncol):
            print ("  {:3.6e}".format(tcoeff[j,l]),end="",file=fout)
        print (" :   1",file=fout)
        #
    #    
    if (fileout != None):
        fout.close()
    return fileout
```
